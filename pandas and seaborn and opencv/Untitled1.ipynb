```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import urllib.request
```

```python
url = "https://im.indiatimes.in/content/2019/Jun/marvel_fans_start_a_petition_to_demand_robert_downey_jr_aka_tony_stark_aka_iron_man_back_1559715390_725x725.jpg"
```

```python
urllib.request.urlretrieve(url , "ironman.jpg")
```

```python
data = cv2.imread("ironman.jpg")
```

```python
plt.imshow(data)
```

```python
rgb = cv2.cvtColor(data , cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb)
```

```python
classifier = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

```python
face = classifier.detectMultiScale(rgb)
```

```python
face
```

```python
x, y, w, h = face[0]
```

```python
ironman = rgb[y:y+h , x : x+ w]
```

```python
plt.imshow(ironman)
```
